#### <b>4.  Make a table of all the producers and directors of each movie. Find the top 3 producers who have produced movies with the highest average RoI? </b>

```python
imdb_df_new.head(3)
```

```python
# Create a dataframe in which we have no null values for the crew

imdb_with_no_null_crews = imdb_df[~imdb_df['crew'].isnull()]
```

```python
len(imdb_with_no_null_crews)
```

```python
# creating a function which returns the list of producers and directors

def get_producers_and_directors(crew_list):
  try:
    #initalise two lists corresponding to producer and director

    producer_list = []
    director_list = []

    #iterate over the list elements

    for elem in crew_list:
      if elem['job'] == 'Producer':
        producer_list.append(elem['name'])

      if elem['job'] == 'Director':
        director_list.append(elem['name'])
    return producer_list, director_list
    
  except:
    return [] , []
```

```python
get_producers_and_directors(imdb_df_new.loc[1, 'crew'])
```

```python
imdb_df_new[['producers', 'directos']] = imdb_df_new.apply(lambda listt : get_producers_and_directors(listt['crew']), axis = 1, result_type = 'expand' )
```
