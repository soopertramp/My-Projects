**4. Make a table of all the producers and directors of each movie. Find the top 3 producers who have produced movies with the highest average RoI?**

```python
no_null_crew = imdb_df_new[~imdb_df_new['crew'].isnull()]
len(no_null_crew)
```

```python
def create_producer_list(index):
  movie_index = no_null_crew.iloc[index]
  crew_list = movie_index.loc['crew']
  producer_list = []

  for elem in crew_list:
    if elem['job'] == 'Producer':
      producer_list.append(elem['name'])
  
  return producer_list
```

```python
create_producer_list(1761)
```

```python
create_producer_list(1757)
```

```python
def create_director_list(index):
  movie_index = no_null_crew.iloc[index]
  crew_list = movie_index.loc['crew']
  director_list = []

  for elem in crew_list:
    if elem['job'] == 'Director':
      director_list.append(elem['name'])
  
  return director_list
```

```python
create_director_list(1745)
```

```python
create_director_list(1758)
```

```python
Table = pd.DataFrame(columns = ['Movie Title', 'Producers', 'Directors', 'ROI'])
```

```python
for index,row in no_null_crew.iterrows():
  try:
    Table = Table.append({'Movie Title' : no_null_crew.loc[index, 'original_title'],
                          'Producers' : create_producer_list(index),
                          'Directors' : create_director_list(index),
                          'ROI' : no_null_crew.loc[index, 'roi']}, ignore_index = True)
    
  except:
    continue
```

**5. Which actor has acted in the most number of movies? Deep dive into the movies, genres and profits corresponding to this actor.**

```python
no_null_cast = imdb_df_new[~imdb_df_new['cast'].isnull()]
```

```python
no_null_cast.loc[0,'cast'][0]['name']
```

```python
actors_list = []

for index,row in no_null_cast.iterrows():
  for iter in no_null_cast.loc[index,'cast']:
    if type(iter) == dict:
      actor = iter['name']
      actors_list.append(actor)
```

```python
actor_table = pd.DataFrame(actors_list, columns = ['Name of The Actor'])
```

```python
actor_table.head()
```

```python
most_no_movies = actor_table.value_counts().reset_index()
most_no_movies.head()
```

```python
print(f"The top 3 actor in this imdb data is {most_no_movies.loc[0, 'Name of The Actor']}, {most_no_movies.loc[1, 'Name of The Actor']}, {most_no_movies.loc[2, 'Name of The Actor']} has acted {most_no_movies.loc[1, ][0]} times, {most_no_movies.loc[1, ][0]} times, {most_no_movies.loc[2, ][0]} times respectively")
```

```python
profit1 = []
profit2 = []
profit3 = []
movie1 = []
movie2 = []
movie3 = []

for index, row in no_null_cast.iterrows():
  for iter in no_null_cast.loc[index, 'cast']:
    if type(iter) == dict:
      actor = iter['name']
      if 'Samuel L. Jackson' in actor:
        profit1.append(no_null_cast.loc[index, 'profit'])
        movie1.append(no_null_cast.loc[index, 'original_title'])

      if 'Robert De Niro' in actor:
        profit2.append(no_null_cast.loc[index, 'profit'])
        movie2.append(no_null_cast.loc[index, 'original_title'])

      if 'Morgan Freeman' in actor:
        profit3.append(no_null_cast.loc[index, 'profit'])
        movie3.append(no_null_cast.loc[index, 'original_title'])
```

```python
#genre

genre_sam = []
a = []

for i in range(len(movie1)):
  for g in no_null_cast.loc[i,'genres']:
    a.append(g['name'])
  
  genre_sam.append(a)
  a = []

genre_rob = []
b = []

for i in range(len(movie2)):
  for g in no_null_cast.loc[i,'genres']:
    b.append(g['name'])
  
  genre_rob.append(b)
  b = []

genre_morg = []
c = []

for i in range(len(movie3)):
  for g in no_null_cast.loc[i,'genres']:
    c.append(g['name'])
  
  genre_morg.append(c)
  c = []
```
